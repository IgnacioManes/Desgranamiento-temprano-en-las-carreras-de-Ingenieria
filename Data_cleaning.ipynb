```python
import pymysql.cursors
import pandas as pd
import numpy as np
```

```python
# Connect to the database
connection = pymysql.connect(host='localhost',
                             user='root',
                             password='',
                             database='TPP',
                             cursorclass=pymysql.cursors.DictCursor)
```

```python
activities = pd.read_sql(("select b.activity_id,b.user_id,status,role_id,a.course_id as course_id from TPP.activities a "   
           "inner join TPP.activity_submissions b on a.id = b.activity_id "
           "inner join TPP.course_users d on b.user_id = d.user_id and a.course_id = d.course_id " 
           "limit 10000000;"), con=connection)
activities.head()
activities = activities[activities["course_id"]>5]
activities = activities[activities["role_id"]==2]
activities["dummy"]=1
```

```python
activities_pivot = activities.pivot_table(index=['activity_id',"course_id",'user_id'], columns='status', aggfunc='count')["dummy"]
activities_pivot["dummy"]=1
```

```python
activities_pivot = activities_pivot.reset_index()
activities_pivot[(activities_pivot['course_id'] == 33)]['user_id'].unique()
```

```python
activities_grouped = activities_pivot.reset_index().groupby(["course_id","user_id"]).count()
```

```python
activities_grouped = activities_grouped.reset_index()
activities_grouped[(activities_grouped['course_id'] == 33)]
```

```python
activities_grouped
```

```python
courses = pd.read_sql(("select * from TPP.courses a "     
           "inner join TPP.course_users b on a.id = b.course_id " 
           "limit 10000000;"), con=connection)
courses.head()
courses = courses[courses["course_id"]>5]
courses["dummy"] = 1
courses = courses[courses["role_id"]==2]
courses["semester"] = courses["semester"].apply(lambda x : x.replace("1c2020","1C-2020").replace("2020-2C","2C-2020").replace("20201C","1C-2020").replace("20201C","1C-2020").replace("2021-1C","1C-2021").replace("2021-2C","2C-2021").replace("2C 2020","2C-2020").replace("2C2020","2C-2020").upper())
```

```python
courses_grouped = courses.groupby(["user_id","course_id","semester","university_course_id"]).count()
courses_grouped = courses_grouped.groupby(["user_id","semester","university_course_id"]).count()["dummy"].rename("courses_per_semester")
```

```python
courses_grouped = courses_grouped.reset_index()
```

```python
users   = pd.read_sql(("select * from TPP.users "
           "limit 10000000;"), con=connection)
users.head()
```

```python
merged_users = activities_grouped.reset_index().merge(users,how='left',right_on ='id',left_on='user_id')
```

```python
merged_users
```

```python
merged_users = merged_users.merge(courses_grouped,how='left', left_on='id',right_on='user_id')
```

```python
merged_users = merged_users.fillna(0)
```
